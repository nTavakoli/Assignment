Assignment 1 Part A: Make a contour plot of the first time step of our dataset

Dataset: Daily Soil Moisture from ESA Climate Change Initiative (CCI) SM v06.1 (Temporal Resolution: 2011-2020, Spatial Resolution: 0.25°, Case study: Global)

```python
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

```python
file = '/scratch/ntavakol/ntavakol/CCI_Final/CCI/ESACCI.nc'
ds = xr.open_dataset(file)
ds
```

```python
con = plt.contourf(ds['sm'][0,-1::-1,:])

# Add colorbar
plt.colorbar(con,shrink=0.7,orientation='horizontal',label='Volumetric Soil Moisture (m3 m-3)')

# Add title
plt.title('ESA CCI SM version, v06.1 (2011-01-01)')
```
